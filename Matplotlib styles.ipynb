```python
import matplotlib.pyplot as plt
import numpy as np
print plt.style.available
```

```python
## Make some data for plotting purposes
N = 100
x = np.random.randn(N)
y = np.random.randn(N)
z = np.arange(N)
def make_plot(**kwargs):
    plt.plot(x,y,'o',**kwargs)
im = np.random.randn(N,N)
def make_img(**kwargs):
    plt.imshow(im,**kwargs)
    plt.colorbar()
```

```python
plt.figure(); make_plot()
plt.figure(); make_img()
```

```python
for style in plt.style.available:
    with plt.style.context(style):
        fig = plt.figure()
        make_plot()
        plt.title(style)
        plt.figure()
        make_img()
        plt.title(style)
```
