```python
## Make some data for plotting purposes
N = 100
x = np.random.randn(N)
y = np.random.randn(N)
z = np.arange(N)
def make_plot(**kwargs):
    plt.plot(x,y,'o',**kwargs)
```

### Styles

You probably don't like the default matplotlib colors or fonts. Here is a simple way to change them.

http://stanford.edu/~mwaskom/software/seaborn/tutorial/aesthetics.html

```python
make_plot()
```

Note that the ipython notebook has a different default style which is better. 

But I like this version more.

```python
import seaborn as sns
make_plot()
```

All we had to do was import seaborn and it automatically applies its styles.

The default colormap has been changed, and the axes style has been changed to a Tufte-like grid.

Seaborn has multiple styles that all look pretty good. You can change the gridding:

```python
for name in ['whitegrid','darkgrid','dark','white','ticks']:
    with sns.axes_style(name):
        fig = plt.figure()
        make_plot()
```

You can change the overall font sizes, line widths, and point sizes:

```python
for name in ['paper','notebook','talk','poster']:
    with sns.plotting_context(name):
        fig = plt.figure()
        make_plot()
```

You can see the full list of possible changes in the documentation of `sns.set`

Typically you're going to want ticks, and for aesthetic or journal restriction reasons you want them to point in instead of out. So I make all my plots with this setting.

The context is "talk", which makes large figures with big font

The style is "ticks"

I like serif font instead of sans-serif.

Of the default color palettes, I like the muted color palette.

So we can put that all together like this:

```python
import seaborn as sns
sns.set(context='talk',style='ticks',font='serif',palette='muted',rc={"xtick.direction":"in","ytick.direction":"in"})
make_plot()
```

Once you've imported seaborn, if you want to reset to the default matplotlib rcParams, you can use `sns.reset_defaults()`. 

The parameters are hardcoded so it restores the cruddy default matplotlib colors.

```python
sns.reset_defaults()
make_plot()
```

### Color Palettes

One of the main things I use seaborn for is to pick color palettes.

Most of the content here is drawn from this tutorial: http://stanford.edu/~mwaskom/software/seaborn/tutorial/color_palettes.html

```python
sns.reset_defaults()
sns.palplot(sns.color_palette()) #current palette
```

```python
for name in ['deep','muted','pastel','bright','dark','colorblind']:
    sns.palplot(sns.color_palette(name))
```

You can also have sequential, diverging, and named colors that are all hardcoded into seaborn. I generally like them more than the current matplotlib defaults.
